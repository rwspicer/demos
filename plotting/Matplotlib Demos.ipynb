```python
n_months= 5 * 12

fig, axs = plt.subplots(3,2,figsize=(12,10))
fig.suptitle('Air Temp', size=30, weight='bold')


axs[0][0].plot(at['dates'][:n_months], at['pdb'][:n_months], color='tab:blue', linestyle='--', label='Prudhoe Bay')
axs[1][0].plot(at['dates'][:n_months], at['utq'][:n_months], color='tab:blue', linestyle='--', label='Utqiaġvik')
axs[2][0].plot(at['dates'][:n_months], at['kak'][:n_months], color='tab:blue', linestyle='--', label='Kaktovik')

axs[0][1].plot(pr['dates'][:n_months], pr['pdb'][:n_months], color='tab:green', linestyle='--', label='Prudhoe Bay')
axs[1][1].plot(pr['dates'][:n_months], pr['utq'][:n_months], color='tab:green', linestyle='--', label='Utqiaġvik')
axs[2][1].plot(pr['dates'][:n_months], pr['kak'][:n_months], color='tab:green', linestyle='--', label='Kaktovik')


for c in range(2):
    for r in range(2):
        axs[r][c].set_xticks([])

for c in range(2):
    axs[2][c].set_xticks(np.arange(0, n_months+1, 12))
    axs[2][c].set_xticklabels(np.arange(1901,1907))
    axs[2][c].tick_params(axis='x', labelrotation=45,labelsize=10)


axs[2][0].set_xlabel('Time',size=20, x=1.1)


axs[1][0].set_ylabel('Temperature ($^\circ$C)',size=20)

axs[1][1].yaxis.set_label_position("right")
for r in range(3):
    axs[r][1].yaxis.tick_right()
axs[1][1].set_ylabel('Precipitation (mm)',size=20)

for i in range(3):
    axs[i][0].set_title(axs[i][0].lines[0].get_label(),x=1.1, y = 1)
```
